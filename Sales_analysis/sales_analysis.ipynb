### Question 1: What was the best month for sales? How much was earned that month?

##### For doing this task, we need a two new columns - month and sales. 
We can achieve Sales by multiplying two existing columns; Quantity Ordered and Price Each.
Then, we need a groupby function and add function to see the monthly sales.

### Task 2: Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2].astype('int64')
all_data
```

### Task 3: Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
plt.bar(months, monthly_sales)
plt.xticks(months)
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

### Task 5: Add a city column

```python
# all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
# all_data.head()
# Or we can define a function
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data
```

### Question 2: What city had the highest number of sales?

```python
results = all_data.groupby('City').sum()
results
```

```python
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.ylabel('Sales')
plt.xlabel('City name')
plt.show()
```

### Question 4: What products are most often sold together?

```python
all_data.head()
```

##### From the table above, we can find index 3 and 4 have the same Order ID number which we can interpret it as sold together.
##### So, let's look through those rows that have the same Order ID number by using `duplicates`

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))  # grouping all products with the same order id in the same line.
df.head()
```

##### Here, we have duplicated rows. We need to drop these by using drop_duplicates

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

##### Now we need to count how many products in each rows.

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

count.most_common(10)
```

```python
for key, value in count.most_common(10):
    print(key, value)
```

### Question 5: What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.figure(figsize=(12, 6))
plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
print(prices)
```

#### How to overlay the average price plot over the existing plot?

##### We can make it by utilizing the `plt.subplots()`!

```python
fig, ax1 = plt.subplots(figsize=(12,6))

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'c')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='c')
ax2.set_ylabel('Average price per unit', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```

#### We can clearly see that the most sold products are Batteries(4-pack) followed by USB-C and Lightning Charging Cables. These items are priced less than $50 so we may say the cheaper it price, the most it sold. We found some interesting data, however, that Laptops and iphones are exceptions in this criteria. Those are the most expensives but sold relatively well against the odds, which can be explained that the demands for these devices have been high during the period.
